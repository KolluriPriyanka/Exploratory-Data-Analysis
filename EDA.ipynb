# 2. The relation Between Specialization and Gender

```python
cross_tab = pd.crosstab(df['Specialization'], df['Gender'])
```

```python
cross_tab
```

```python
# Create a stacked bar chart to visualize the relationship between gender and specialization
cross_tab.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Specialization')
plt.ylabel('Count')
plt.title('Gender Distribution by Specialization')
plt.legend(title='Gender')
plt.show()
```

### Statistical Analysis

```python
from scipy.stats import chi2_contingency

# Perform chi-squared test for independence
chi2, p_value, _, _ = chi2_contingency(cross_tab)

# Check if p-value is below a significance level (e.g., 0.05) to determine if there is a significant relationship between gender and specialization
significance_level = 0.05
if p_value < significance_level:
    print("There is a significant relationship between Gender and Specialization.")
else:
    print("There is no significant relationship between Gender and Specialization.")
```

## There is a significant relationship between Gender and Specialization.The female are very LOW in all the specialization

# Feature Transformation

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder  #importing required classes for Transformation
```

```python
# Drop rows where JobCity is equal to '-1'
df1= df[df['JobCity'] != ' -1']   #removing the unnecessary characters
```

```python
# One-Hot Encoding for categorical columns
cat_cols = ['10board', '12board', 'Degree', 'Specialization', 'CollegeState']
one_hot_encoder = OneHotEncoder()
df_one_hot = pd.get_dummies(df, columns=cat_cols)

# Standard Scaling for numerical columns
num_cols = ['Salary', '10percentage', '12percentage', 'collegeGPA',
            'English', 'Logical', 'Quant', 'Domain', 'ComputerProgramming', 'ElectronicsAndSemicon',
            'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg', 'conscientiousness',
            'agreeableness', 'extraversion', 'nueroticism', 'openess_to_experience', 'Age']
scaler = StandardScaler()
df_one_hot[num_cols] = scaler.fit_transform(df_one_hot[num_cols])

# Label Encoding for remaining categorical columns
label_encoder = LabelEncoder()
for col in df_one_hot.columns:
    if df_one_hot[col].dtypes == 'object':
        df_one_hot[col] = label_encoder.fit_transform(df_one_hot[col].astype(str))
```
